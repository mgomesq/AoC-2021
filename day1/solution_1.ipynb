```python
file = open("puzzle_input.txt")
```

```python
lista = file.readlines()
```

```python
lista
```

```python
opa_lista = []
for y in lista:
	opa = y.replace("\n","")
	opa = int(opa)
	opa_lista.append(opa)
print(opa_lista)
```

```python
tam = len(opa_lista)
x = 1
valor_pos = 0
while x < tam:
	valor_novo = opa_lista[x]
	valor_velho = opa_lista[x-1]
	delta = valor_novo - valor_velho
	x = x + 1
	if delta > 0:
		valor_pos = valor_pos + 1
print("Número de vezes que aumentou: ",valor_pos)
```
