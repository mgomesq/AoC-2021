```python
file = open('puzzle_input.txt')
input_list = file.readlines()
input_list = [x.replace('\n', '') for x in input_list]
```

```python
#--- Part One ---
horizontal_position = 0
depth = 0

for instruction in input_list:
	command, times = instruction.split(' ')
	times = int(times)
	if command == 'forward':
		horizontal_position += times
	if command == 'down':
		depth += times
	if command == 'up':
		depth -= times

print('Horizontal position: ', horizontal_position)
print('Depth: ', depth)

print('Solution: ', horizontal_position*depth)
```

```python
#--- Part Two ---

aim = 0
horizontal_position = 0
depth = 0

for instruction in input_list:
	command, times = instruction.split(' ')
	times = int(times)
	if command == 'forward':
		horizontal_position += times
		depth += aim*times
	if command == 'down':
		aim += times
	if command == 'up':
		aim -= times

print('Horizontal position: ', horizontal_position)
print('Depth: ', depth)

print('Solution: ', horizontal_position*depth)
```
