```python
with open('puzzle_input.txt') as file:
	input_list = file.readlines()
	
input_list = [element.replace("\n", "") for element in input_list]
```

```python
# Change of axis

# Assuming every element is of the same length
binary_lenght = len(input_list[0])
input_counts = len(input_list)

binary_positions =[[] for _ in range(binary_lenght)]

for element in input_list:
	for idx, char in enumerate(element):
		binary_positions[idx].append(int(char))

position_sum = [sum(position) for position in binary_positions]
```

```python
threshold = int(input_counts/2)

gamma_rate = ['1' if bit >threshold else '0' for bit in position_sum]
epsilon_rate = ['0' if bit == '1' else '1' for bit in gamma_rate]
```

```python
gamma_rate = int(''.join(gamma_rate), 2)
epsilon_rate = int(''.join(epsilon_rate), 2)
```

```python
power_consumption = gamma_rate*epsilon_rate
```
